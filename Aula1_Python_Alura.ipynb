```python
linhas, colunas = df.shape[0], df.shape[1]
print("linhas:", linhas)
print("colunas:", colunas)
```

```python
df.columns
```

```python
renomear_colunas = {
    "ano_de_trabalho": "ano_trabalho",
    "nivel_de_experiencia": "nivel_experiencia",
    "tipo_de_emprego": "contrato",
    "cargo": "cargo",
    "salario": "salario",
    "moeda_do_salario": "moeda",
    "salario_em_dolar": "moeda_usd",
    "residencia_do_empregado": "residencia_empregado",
    "horas_de_trabalho_remoto": "remoto",
    "localizacao_da_empresa": "localizacao_empresa",
    "tamanho_da_empresa": "tamanho_empresa"
}

df.rename(columns = renomear_colunas, inplace=True)
df.columns
```

```python
df["nivel_experiencia"].value_counts()
#SE = senior / MI = intermediário(pleno) / EN = entry (iniciante) / EX = executivo
```

```python
df["contrato"].value_counts()
#FT = fulltime / CT = contrato temporário / PT = part-time(meio período) / FL = freelancer
```

```python
df["remoto"].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
#M= médio / L =large / S = small
```

```python
traducao_tamanho = {
    "M": "Médio",
    "L": "Grande",
    "S": "Pequeno"
}

df["tamanho_empresa"] = df["tamanho_empresa"].replace(traducao_tamanho)
df["tamanho_empresa"].value_counts()
```
